```python
data=sns.load_dataset("tips")    # load dataset form seaborn
```

```python
data.head()
```

```python
data.tail()
```

```python
plt.plot(data['total_bill'])
```

```python
sns.lineplot(data['total_bill'])   #line plot for total bill
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)    #we are doing scatter plot of total bill and tip
```

```python
sns.histplot(data["total_bill"])    # hist of total_bill
```

```python
plt.hist(data["total_bill"])
```

```python
sns.kdeplot(data["total_bill"])
```

```python
#catplot
#relplot  #scatterplot
```

```python
sns.relplot(x="total_bill",y='tip',data=data)     #we are doing scatter plot of total bill & tip
```

```python
data['time'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='time')      #
```

```python
sns.relplot(x='total_bill', y='tip',hue='time',style='smoker',data=data) 
# we are doing scatter  plot of total_bill and tip
```

```python
sns.relplot(x='total_bill', y='tip',hue='time',style='smoker',data=data,kind='scatter',col='smoker')    # col divide of bases of unique value
```

```python
sns.relplot(x='total_bill', y='tip',hue='time',style='smoker',data=data,kind='scatter',row='time') 
```

```python
data['day'].unique()
```

```python
sns.relplot(x='total_bill', y='tip',hue='time',data=data,kind='scatter',col='day')   # graph at particular day 
```

```python
sns.relplot(x='total_bill', y='tip',data=data) 
plt.title('total_bill vs tip')
plt.xlabel('total_bill',fontsize=30,color='red')
plt.ylabel('tip',fontsize=30,color='red')
plt.show()
```
