## Modelling

1 confusin matrix  (tp(model and actual postive),tn(model=neg ,actual=neg),fn(model=neg,actual=pos),fp(model=pos,actual=neg))
false postive is reduce

2 accuracy= (tp+tn)/(tp+tn+fn+fp)

3 precission=  (tp)/(tp+fp)

4 recall  =  (tp)/(tp+fn)

5 f1score  = (2*precision*recall)/(precision+recall)

fp type 2 error
fn type 1 error
