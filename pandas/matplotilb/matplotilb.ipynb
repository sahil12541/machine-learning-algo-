```python
x=np.array([1,5,9,3,5,4,4,2,7,3])
y=np.array([7,5,3,1,5,9,4,8,5,7])
plt.scatter(x,y)
plt.show()
```

```python
x1=np.array([1,5,9,3,5,4,4,2,7,3])             #comparison b/w two scatter
y1=np.array([7,5,3,1,5,9,4,8,5,7])
x2=np.array([8,5,2,69,4,47,2,4,5,5])
y2=np.array([1,5,9,3,5,4,4,2,7,3])
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.show()
```

```python
x=np.array([1,5,9,3,4,8])             #comparison b/w two scatter
y=np.array([7,5,3,1,2,5])
c=np.array(['red','orange','blue','yellow','black','pink'])       # c is used to give different colour to each lable
plt.scatter(x,y, c=c)
plt.show()
```

```python
x=np.array([1,5,9,3,4,8])             
y=np.array([7,5,3,1,2,5])
size=np.array([500,522,455,112,114,164])      # s is used to give the differet size
plt.scatter(x,y, s=size)
plt.show()
```

```python
x=np.array([1,5,9,3,4,8])             
y=np.array([7,5,3,1,2,5])
size=np.array([500,522,455,112,114,164])      # s is used to give the differet size
plt.scatter(x,y, s=size,alpha=0.5)
plt.show()
```

```python
x=np.array([1,5,9,3,4,8])             
y=np.array([7,5,3,1,2,5])
size=np.array([500,522,455,112,114,164])
c=np.array(['red','orange','blue','yellow','black','pink'])
plt.scatter(x,y, s=size,c=c ,alpha=0.5)      #alpha used for intensity b/w 0 & 1
plt.show()
```

```python
x=np.array([1,5,9,3,4,8])             
y=np.array([7,5,3,1,2,5])
size=np.array([500,522,455,112,114,164])
c=np.array(['red','orange','blue','yellow','black','pink'])
plt.scatter(x,y, s=size,c=c ,alpha=0.5,cmap="nipy_spectral")
#cap=virdis , nipy_spectral
plt.colorbar()       #colorbar used to show colorline outside the graph
plt.show()
```

PIECHAET

```python
y=np.array([1,2,3,4])      # to draw piechart
plt.pie(y)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]      #mylabel give different label of each data 
plt.pie(y,labels=mylabel)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
plt.pie(y,labels=mylabel,startangle=90)   #startangle used to shift 90 degree (starting)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexploed=[0,2,0,0,]
plt.pie(y,labels=mylabel,startangle=90,explode=myexploed)      #exclude is used to move out a part from oie chart
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexploed=[0,2,0,0,]
plt.pie(y,labels=mylabel,startangle=90,explode=myexploed,shadow=True)   #shodow for 2d print 
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexploed=[0,2,0,0,]
mycolor=["green","yellow","red","BLUE"]       # autopct used to divide in persentage ratio
plt.pie(y,labels=mylabel,startangle=90,explode=myexploed,colors=mycolor,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexploed=[0.2,0,0,0]
mycolor=["green","yellow","red","BLUE"]          #legend used to show color outside piechart
plt.pie(y,labels=mylabel,startangle=90,explode=myexploed,colors=mycolor,autopct='%1.1f%%',shadow=True)
plt.legend(title='color')
plt.show()
```
