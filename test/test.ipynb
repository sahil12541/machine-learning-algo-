```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating the DataFrame from the given dataset
data = {
    'Name': ['John', 'Alice', 'Bob', 'Emma', 'Ryan'],
    'Age': [28, 24, 30, 26, 35],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Salary': [50000, 60000, 55000, 62000, 72000],
    'Department': ['IT', 'HR', 'Finance', 'IT', 'Finance']
}

df = pd.DataFrame(data)

# 17. Calculate the average salary using NumPy
average_salary = np.mean(df['Salary'])
print(f"Average Salary: {average_salary}")

```

```python
# 18. Filter out employees with a salary greater than 60,000 using Pandas
high_salary_employees = df[df['Salary'] > 60000]
print("\nEmployees with salary greater than 60000:")
print(high_salary_employees)
```

```python
# 19. Plot a bar chart for department-wise average salary using Matplotlib
dept_avg_salary = df.groupby('Department')['Salary'].mean()

plt.figure(figsize=(8, 5))
plt.bar(dept_avg_salary.index, dept_avg_salary.values, color=['blue', 'green', 'red'])
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Department-wise Average Salary")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 40],
    'Department': ['HR', 'IT', 'Sales', 'HR', 'IT'],
    'Salary': [50000, 70000, 45000, 52000, 80000]
}
df = pd.DataFrame(data)

# Problem 20: Bar chart of number of employees per department
department_counts = df['Department'].value_counts()
plt.figure(figsize=(8, 5))
department_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees in Each Department')
plt.xticks(rotation=0)
plt.show()




```

```python
# Problem 21: Filter employees with salary > 60,000
high_salary_df = df[df['Salary'] > 60000]
print("Employees with salary greater than 60,000:")
print(high_salary_df)
```

```python
# Problem 22: Line graph for salary trend by department
plt.figure(figsize=(8, 5))
for department in df['Department'].unique():
    dept_data = df[df['Department'] == department]
    plt.plot(dept_data['Name'], dept_data['Salary'], marker='o', linestyle='-', label=department)

plt.xlabel('Employee')
plt.ylabel('Salary')
plt.title('Salary Trend by Department')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    'Student Name': ['John', 'Sarah', 'David', 'Emily'],
    'Math': [80, 95, 70, 85],
    'Science': [75, 85, 65, 90],
    'English': [90, 80, 75, 88],
    'History': [70, 85, 80, 92]
}

df = pd.DataFrame(data)

# 23. Mean score of each subject
mean_scores = df[['Math', 'Science', 'English', 'History']].mean()
print("Mean Scores:\n", mean_scores)
```

```python
# 24. Average score of each student
df['Average Score'] = df[['Math', 'Science', 'English', 'History']].mean(axis=1)
print("\nAverage Score of Each Student:\n", df[['Student Name', 'Average Score']])
```

```python
# 25. Student with the highest overall score
top_student = df.loc[df['Average Score'].idxmax(), 'Student Name']
print("\nTop Student:", top_student)
```

```python
# 26. Scatter plot for Math vs Science
plt.scatter(df['Math'], df['Science'], color='blue')
plt.xlabel('Math Scores')
plt.ylabel('Science Scores')
plt.title('Math vs Science Scores')
plt.show()
```

```python
# 27. Standardizing scores
standardized_scores = (df[['Math', 'Science', 'English', 'History']] - mean_scores) / df[['Math', 'Science', 'English', 'History']].std()
print("\nStandardized Scores:\n", standardized_scores)



```

```python
# 28. Bar plot for total scores
df['Total Score'] = df[['Math', 'Science', 'English', 'History']].sum(axis=1)
plt.bar(df['Student Name'], df['Total Score'], color=['red', 'green', 'blue', 'orange'])
plt.xlabel('Student Name')
plt.ylabel('Total Score')
plt.title('Total Scores of Students')
plt.show()
```

```python
# 29. Pie chart for subject-wise average scores
plt.pie(mean_scores, labels=mean_scores.index, autopct='%1.1f%%', colors=['yellow', 'blue', 'green', 'red'])
plt.title('Average Subject Scores')
plt.show()
```

```python

# 30. Students with History score greater than average history score
average_history_score = mean_scores['History']
high_history_students = df[df['History'] > average_history_score][['Student Name', 'History']]
print("\nStudents with History score greater than average:\n", high_history_students)
```
